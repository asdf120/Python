# 이변량양적자료

> 1. 이변량 - 변수 2개

> 2. 양적자료 -   숫자형 자료         (ex. 온도, 키, 몸무게 )


            * 질적자료 = 명목형 = 이산형
            * 양적자료 = 수치형 = 연속형

> 3. 다변량 - 변수 여러개

```python
# 데이타 읽어오기
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns                


# 인덱스가 2번 생성되기에 기존 인덱스를 인덱스로 지정
diamonds = pd.read_csv('data/diamonds.csv', index_col='Unnamed: 0')
diamonds
```

```python
# 캐럿(무게)와 가격에 대한 그래프 -> 산점도
data =pd.crosstab(diamonds['price'],diamonds['carat'])
# plt.plot(data=data)
sns.scatterplot(x='price', y='carat', data=diamonds)
```
