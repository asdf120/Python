```python
# 기본적인 그래프

import matplotlib.pyplot as plt

data = [10,15,19,30,20]
plt.plot(data);
```

```python
# 1- 그래프 틀(영역)을 만들기
plt.figure(figsize=(20,10))
# 2- 그래프 그리기
plt.plot(data)
# 3- 그래프 보여주기
plt.show()
```

```python
# x축과 y축의 라벨 추가
# 1- 그래프 틀(영역)을 만들기
plt.figure()
# 2- 그래프 그리기
plt.plot(data)
plt.title('sample')
plt.xlabel('time')
plt.ylabel('temperature')
# 3- 그래프 보여주기
plt.show();
```

1- 그래프 틀(영역)을 만들기
   > plt.figure()       
    
2- 그래프 그리기
   > plt.plot(data)     
    
3- 그래프 보여주기
   > plt.show()         

 
 ### *그래프 출력시 객체표시 같은 거 안 출력하게 하려면 뒤에 세미콜론(;) 붙이면 된다*

```python
# y축 데이타에 x축 데이타를 지정
data = [10,15,19,30,20]
step = ['one','two','three','four','five']
# plt.plot(data,step)
plt.plot(step,data)
plt.grid()
```

```python
# 속성 : linestyle , color , marker
plt.plot(step,data,linestyle='dashed',color='blue',marker='x')
```

```python
# 속성 축약
plt.plot(step,data,'b--x');
```

[참고] 선의 두께와 선의 색을 지정

- lw의 기본값은 1


- 선 색
    {'b','g','r','c','m','y', 'k', 'w' }

```python
plt.plot(step,data,linestyle='dashed',color='blue',marker='x',
        lw = 3)
```

```python
# 마커속성 
plt.plot(step,data,linestyle='dashed',color='blue',marker='*',
        lw=3, markerfacecolor='red', markersize=12)
```
