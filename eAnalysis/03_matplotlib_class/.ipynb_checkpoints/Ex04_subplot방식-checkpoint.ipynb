```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

""" 1.  figure(그림틀=공간?)을 만들고 거기에 subplot을 추가 하는 방식 """

plt.plot(np.random.randn(100))
plt.plot(np.random.randn(200).cumsum());
```

# 하나의 화면에 여러 개의 그래프 그려보기

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(np.random.randn(100));
ax2.plot(np.random.randn(200).cumsum());
```

```python
# x, y 축 값 범위 지정
fig = plt.figure()
ax = fig.add_subplot(111) # 1,1,1 안써도 됨
ax.plot(np.random.randn(100),'b*--');

# 제목, x라벨, y라벨 지정
plt.title('data')
plt.xlabel('data-x')
plt.ylabel('data-y')

ax.set_xticks([10,50,100,150])
# plt.xticks([10,50,100])
```

```python
""" 2. figure로 공간을 만들고 파이플롯(pyplot)의 subplot()으로 각각의 패널 지정 """
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(np.random.randn(100));
ax2.plot(np.random.randn(200).cumsum());

fig = plt.figure()
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.plot(np.random.randn(100));
ax2.plot(np.random.randn(200).cumsum());
```

```python
""" 3. 간단하게 ax 객체를 이용하기 """
fig,ax = plt.subplots(1,2)
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(200).cumsum())
ax = plt.subplots(1,2)
```

## 서브플롯 개념

```python
plt.figure(figsize=(16,8))

plt.subplot(2,2,1) # 2행 2열 첫번째 영역
plt.subplot(222) # 2행 2열 두번째 영역
plt.subplot(212) # 2행 1열 2번째 영역

plt.show()
```

```python
plt.figure(figsize=(16,8))

plt.subplot(2,1,1)
plt.subplot(223)
plt.subplot(224)
```

```python
plt.figure(figsize=(16,8))

plt.subplot(2,3,(1,2))
plt.subplot(2,3,3)
plt.subplot(2,1,2)
```
