```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
plt.style.use('classic')
plt.style.use('seaborn-whitegrid')
x = np.linspace(0,10,100) # 0부터 10까지 수 중에 100개 균등하게 추출
x
# plt.plot(np.sin(x))
```

```python
""" 1. 색상지정 """
plt.plot(np.sin(x-1), color='red');
plt.plot(np.sin(x-2), color='blue');
plt.plot(np.sin(x-3), color='purple');
plt.plot(np.sin(x-4), color='brown');
plt.plot(np.sin(x-5), color='r');
plt.plot(np.sin(x-6), color='#FF8822');
plt.plot(np.sin(x-7), color='CHARTREUSE');
plt.plot(np.sin(x-8), color='0.75'); # 회색 0 ~ 1
plt.plot(np.sin(x-9), color=(.9,1,.3)); # 0~1 사이의 수로 색상 지정 (자료형 튜플)



```

```python
""" 2. 선스타일 """
# plt.plot(np.sin(x), linestyle='solid')
# plt.plot(np.sin(x-1), linestyle='dashed')
# plt.plot(np.sin(x-2), linestyle='dashdot')
# plt.plot(np.sin(x-3), linestyle='dotted')

plt.plot(np.sin(x), linestyle='-')
plt.plot(np.sin(x-1), linestyle='--')
plt.plot(np.sin(x-2), linestyle='-.')
plt.plot(np.sin(x-3), linestyle=':')
```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """

plt.plot(np.sin(x-1),'g-')
plt.plot(np.sin(x-2), '--c')
plt.plot(np.sin(x-3), '-.k')
plt.plot(np.sin(x-4), ':r')
```

```python
''' 4. 범례 '''
plt.plot(np.sin(x-1),'g-', label='first')
plt.plot(np.sin(x-2), '--c', label='second')
plt.plot(np.sin(x-3), '-.k', label='third')
plt.plot(np.sin(x-4), ':r', label='fourth')

plt.legend(loc=1) # loc default 1
```
