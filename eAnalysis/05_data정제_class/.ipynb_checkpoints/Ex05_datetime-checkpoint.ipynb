## datetime 라이브러리 

- date 
- time 
- datetime
    

   > [  df  ]   
 문자열 처리를 위해 str 접근자를 사용하듯이 datetime 오브젝트는 dt 접근자 사용해야 
 datetime 속성이나 메소드를 사용가능
 
     [도서] Do it 데이타분석을 위한 판다스 입문

```python
""" datetime 라이브러리 : date / time / datetime 오브젝트 """
from datetime import datetime

print(datetime.now())
print(datetime.today())

today = datetime.now()
print(today.year)
print(today.month)
print(today.day)

aday = datetime(2019,1,1)
print(today-aday)
# print(aday + 10) # date타입은 int타입과 연산이안됨

print(today.strftime('%Y-%m-%d'))
```

## [ 연습 ]  datatime  오브젝트로 변경하여 필요한 데이타 추출 및 계산

```python
import pandas as pd


ebola = pd.read_csv('./data/ebola_timeseries.csv')
ebola.head()
```

```python
# datetime 오브젝트로 변경하여 필요한 데이타 추출 및 계산
# ebola.info()

ebola['date_dt']=pd.to_datetime(ebola['Date'])
ebola.head()
```

```python
ebola[['Date','date_dt']]
ebola['year'] = ebola['date_dt'].dt.year
ebola[['Date','date_dt','year']]
```
