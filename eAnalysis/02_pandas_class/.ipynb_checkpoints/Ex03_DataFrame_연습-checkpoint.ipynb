#### 판다스 데이타프레임의 기본적인 정보 얻어오기

- 앞에 10개 행 보기
- 뒤에 10개 행 보기

- 판매건수에서 큰 값 순으로 10개 (nlargest()이용)
- 판매건수에서 작은 값 순으로 10개 (nsmallest()이용)


```python
df['판매건수'].nlargest(10)
df.nlargest(10,'판매건수')
```

```python
df.nsmallest(10,'판매건수')
```

#### 행에서 데이타 추출

- 야채, 딸기, 봉투행을 추출하여 매출순으로 정렬하기 (sort_values()이용)
- 바나나행부터 딸기행까지 추출
- 야채부터 딸기행까지 추출

```python
df.loc[['야채','딸기','봉투']].sort_values(by='매출')
df.loc[['야채','딸기','봉투']].sort_values('매출')
```

```python
df.loc['야채':'딸기']
```
