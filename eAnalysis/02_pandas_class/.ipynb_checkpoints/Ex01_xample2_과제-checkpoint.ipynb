### * 인사이트를 도출하여 정보를 추출하고 시각화

## 역대 대통령들의 이전 직업

```python
import matplotlib.pyplot as plt
occupation_num = df['Occupation'].value_counts()
plt.figure(figsize=(10,5))
occupation_num.plot(kind='bar')
```

## 대통령 취임 나이

```python
inaugration_age = df['Age at inauguration'].value_counts()
inaugration_age
plt.figure(figsize=(10,9))
inaugration_age.plot(kind='barh')
```



## 역대 대통령들의 선출된 주 비율

```python
state_rating = df['State elected from'].value_counts()
state = ['Ohio','New York','Virginia','Illinois','Messachusetts','Texas','Tennessee','California',
        'Arkansas','Indiana','New Hampshire','New Jersey','Pennsylvania','Louisiana','Georgia']
plt.figure(figsize=(10,10))
plt.pie(state_rating,labels=state,autopct='%.1f%%')
plt.title('State elected from')
plt.show()
```

## 선출된 정당의 비율

```python
party_rating = df['Political Party'].value_counts()
party = ['Republican','Democrat','Democratic-Republican','Whig','None','Federalist']
plt.figure(figsize=(10,10))
plt.pie(party_rating,labels=party,autopct='%.1f%%')
plt.title('Political Party')
plt.show()
```

## 출신 대학

```python
college_rating = df['College'].value_counts()
college_count = []
etc_count = 0
for count in college_rating:
    if count == 1:
        etc_count += 1
    else :
        college_count.append(count)
college_count.append(etc_count)
college = ['None','Harvard','Yale','Princeton','William and Mary','US Military Academy','etc']
plt.figure(figsize=(10,10))
plt.pie(college_count,labels=college,autopct='%.1f%%')
plt.title('College')
plt.show()
```
